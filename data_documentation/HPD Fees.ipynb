## Documentation

### Attribution:
New York City Department of Buildings (DoB)

### Semantic Contents 

#### Description:
This is a dataset of fees charged to building owners as reported by the Department of Buildings
#### Selection:
A BBL was generated from the respective columns and the following columns were kept or generated based on a BBL GROUP BY:
- TotalFeeAmount: sum of fees (int)
- NumberOfComplaints: count of instances (int)
- ModeFeeTypeName: mode of type of fee (string)
- ModeFeeTypeCount: count in mode (int)
- ModeLifeCycleName: mode of building cycle (string) 
- ModeLifeCycleCount: count in mode (int)
- MeanFeeIssueDate: average date fee issued (int)
#### Date timeframe:
2008-2019

### Collection Process
Collected from the NYC Open Data project. 
#### How the data was gathered:
https://data.cityofnewyork.us/Housing-Development/Fee-Charges/cp6j-7bjj
