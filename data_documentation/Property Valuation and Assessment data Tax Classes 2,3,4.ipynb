## Documentation

### Attribution:
Department of Finance (DOF)
[Link to source (as of 3-31-19)](https://data.cityofnewyork.us/Housing-Development/Property-Valuation-and-Assessment-data-Tax-Classes/cqds-77ys)

### Semantic Contents 
#### Description:
Descriptive building data, such as size and number of units, as well as data used by the department of finance for property tax. 

#### Selection:

- BBL- constructed from constituent elements: str	
- Owner- owner listed on dead or title: str
- Total_Units- total number of units (any kind): int	
- Residential_Units	total number of residential units: int
- Buildings- number of buildings on the lot: int
- Stories- highest number of building stories on the lot: int	
- Land_Market_Value- market value of the lot's land
- Total_Market_Value- total market value of the lot and its building assets	
- Transitional_Assesed_Land_Value- transition assesed value. According to the NYC department of finance: "We phase in changes to the assessed value of Class 2 properties with more than 10 units and Class 4 properties over a five-year period. This means that we apply 20% of the change each year for five years.  "In any given year, there are multiple transitions being applied which results in an actual assessed value and a transitional assessed value for your property each year. The law requires that we use whichever number is lower – your actual assessed value or transitional assessed value – to get to your property tax bill."  NYC DoF (https://www1.nyc.gov/site/finance/taxes/property-determining-your-transitional-assessed-value.page)	
- Transitional_Assessed_Total_Value- transitional land value (see above) 	
- Actual_Assesed_Land_Value	- (see above)
- Actual_Assessed_Total_Value - (see above)
- 5_Year_Exemption_Flags- Between zero and 5 'E' to represent number of exemption years in the past 5.	
- Exempt_Total_Value - Value of exemptions
