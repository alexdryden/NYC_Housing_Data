### Semantic Contents 
#### Description:
This is the table representing the problems associated with complaints

#### Selection:
Values were grouped by building(BBL):
- BBL-constructed from constituent columns: datatype= str
- MeanReceivedDate-Mean of dates for complaints per building: datatype= datetime
- ApartmentComplaints- count, constructed from the UnitType column: datatype=float
- BuildingComplaints- count, constructed from the UnitType column	: datatype=float
- EMERGENCY- count, constructed from the type column	: datatype=float
- HAZARDOUS- count, constructed from the type column: datatype=float
- IMMEDIATE_EMERGENCY- count, constructed from the type column	: datatype=float
- NON_EMERGENCY- count, constructed from the type column	: datatype= float
- TotalComplaints- count, calculated with groupby on the constructed BBL column : datatype= int
#### Date timeframe:
2013-2019
